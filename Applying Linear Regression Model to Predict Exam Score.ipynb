# **Applying Linear Regression Model For Predicting Exam Score**

```python
#Importing Packages and libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Reading CSV File
df = pd.read_csv("/kaggle/input/student-study-hours/score_updated.csv")

#Checkng Colummn Names and Format
df.head()
```

```python
#Plotting Datapoints in Dataset
plt.figure(figsize=(10, 6))
plt.scatter(df['Hours'], df['Scores'], c = 'black' , marker = 'o')

#Adding Title and Labels
plt.title('Hours vs Score')
plt.xlabel('Score')
plt.ylabel('Hours')

#Display 
plt.show()
```

```python
#Creating Array for X
x = np.array(df['Hours'])

#Creating Array for Y
y = np.array(df['Scores'])
```

```python
#Creating Gradient Dscent Function
def Gradient_Descent(w, b, x, y, L):
    n = len(x)
    for i in range(n):
                   linear = w * x[i] + b
                   mean_error = (linear - y[i]) 
                   w -= L * (mean_error * x[i])
                   b -= L * mean_error
    return w, b
```

   **Linear Regression Implementaion**

```python
#Initializing Values
w = 0
b = 0
L = 0.01
iterations = 1000

for i in range(iterations):
    w, b = Gradient_Descent(w, b, x, y, L)
    
#Creatimg Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c='black', marker = 'o')
plt.plot(df['Hours'], w * df['Hours'] + b, c='blue')

#Displaying Graph
plt.show()

```
